```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
n = 1000
noise = np.random.default_rng().normal(loc=0.0, scale=0.05, size=(2,n))
```

```python
radians = np.random.rand(n) * 2 * np.pi
(xs, ys) = np.cos(radians), np.sin(radians)
```

```python
plt.scatter(xs + noise[0],ys + noise[1])
```

```python
#make labels
labels = (1 * ((xs > 0) & (ys < 0)) 
          + (np.random.default_rng().binomial(1, 0.25, n)) * (ys >= 0) 
         )
```

```python
colors = ['b' if label else 'r' for label in labels]
plt.scatter(xs + noise[0],ys + noise[1], color =colors)
```

```python
u, c = np.unique(labels, return_counts=True)
```
